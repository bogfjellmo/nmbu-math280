```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python

e=np.exp(1)

c0=4*e-10
c1=18-6*e
print(c0,c1)
```

```python

t=np.linspace(0,1)
pt= c0+c1*t
plt.plot(t, np.exp(t), t, pt)
```

```python
A= np.array([[1, 1/2],[1/2, 1/3]])
b = np.array([np.exp(1)-1, 1])
c=np.linalg.solve(A,b)
print(c)
```
