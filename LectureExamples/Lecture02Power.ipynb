```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A= np.array([[1.2, -0.5], 
             [0.15, 1.8]])
print(A)
```

```python
v0=np.array([1,0])
print(v0)
```

```python
2*v0
```

```python
2*A
```

```python
A@v0
```

```python
A.dot(v0)
```

```python
v=v0
vlist=[v]
M=30
for _ in range(M):
    v=A.dot(v)
    idx = np.argmax(np.abs(v))
    v/=v[idx]
    vlist.append(v)
```

```python
fig, ax = plt.subplots()
ax.set(xlim=(-3, 3), ylim=(-1,3))
for v in vlist:
    ax.plot([0, v[0]], [0,v[1]])
ax.legend(list(range(M+1)))   
```
