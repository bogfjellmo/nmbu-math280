# A digit data set
[Scikit-learn](https://scikit-learn.org/) comes with a dataset of handwritten digits. [Link](https://scikit-learn.org/stable/datasets/toy_dataset.html#digits-dataset)


```python
from sklearn.datasets import load_digits
digits=load_digits()

plt.matshow(digits.images[0], cmap='gray')
```

```python
digits.data[0]
```

```python
X=digits.data
C=digits.target
X.shape
```

```python
#Pick out the first 100 images

X=X[0:100,:]
C=C[0:100]

X=X.T
X.shape
```

```python
plt.figure()
plt.axis()
plt.scatter(X[1,:], X[2,:], c=C, cmap='tab10')
plt.colorbar()
```

```python
# compute the eigenvalues and eigenvectors
A=X@X.T
l,v =np.linalg.eig(A)
```

```python
u1=v[:,0] 
u2=v[:,1]
u3=v[:,2]
```

```python
y1=u1@X
y2=u2@X
y3=u3@X
```

```python
plt.figure()
plt.scatter(y1, np.zeros_like(y1), c=C, cmap='tab10')
plt.colorbar()
plt.title('1D-projection')
```

```python
plt.figure()
plt.scatter(y1, y2, c=C, cmap='tab10')
plt.colorbar()
plt.title('2D-projection')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plot=ax.scatter(y1,y2, y3, c=C, cmap='tab10')
ax.set_title('3D-projection')
```
