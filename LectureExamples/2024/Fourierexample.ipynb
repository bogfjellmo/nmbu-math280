```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=100
t=np.linspace(0,2*np.pi,N+1)[:-1] # get N points in [0, 2pi).

f=1+np.sin(t)-3*np.cos(5*t)+2*np.sin(15*t)
```

```python
plt.plot(t,f)
```

```python
fhat=np.fft.rfft(f)
```

```python
plt.plot(np.abs(fhat))
```

```python
fhat.round(4)
```
