```python
import numpy as np
```

```python
P=np.zeros((5,5))
P[1,0]=1/3
P[2,0]=1/3
P[3,0]=1/3
P[4,1]=1
P[1,2]=1
P[2,3]=1
P[0,4]=0.5
P[3,4]=0.5
print("P = \n" + str(P))
```

```python
x=np.array([1,0,0,0,0])
for i in range(21):
    print("x_"+str(i)+" =\t"+str(x))
    x=P@x
```
