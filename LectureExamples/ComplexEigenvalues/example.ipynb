```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
u=np.array([1,1])
v=np.array([0,1])

a=np.cos(1)
b=np.sin(1)
```

```python
P=np.stack((u,v)).T
C=np.array([[a,-b],[b,a]])

A=P@C@np.linalg.inv(P)
print(A)
```

```python
np.linalg.eig(A)
```

```python
x0=np.array([1,0])
x=x0
```

```python
plt.plot(x0[0], x0[1], 'o')
plt.xlim(-2.1,2.1)
plt.ylim(-2.1,2.1)
```

```python
x=A@x
plt.plot(x[0],x[1], 'o')
```
