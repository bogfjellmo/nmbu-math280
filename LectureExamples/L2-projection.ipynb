```python
from sympy import *

t=Symbol('t')


p0=1
p1=t
p2=t**2


def inner(f,g):
    return integrate(f*g,(t,0,1))

def norm(f):
    return sqrt(inner(f,f))
```

```python
# Gram schmidt

q0=p0
q1=p1-inner(p1,q0)/inner(q0,q0)*q0
q2=p2-inner(p2,q0)/inner(q0,q0)*q0-inner(p2,q1)/inner(q1,q1)*q1
```

```python
display(q0)
display(q1)
display(q2)
```

```python
plot(q0/norm(q0),q1/norm(q1),q2/norm(q2),(t,0,1))
```

```python
display(inner(q0, exp(t)))
display(inner(q1, exp(t)))
display(inner(q2, exp(t)))
```

```python
def proj(f):
    return inner(f,q0)/inner(q0,q0)*q0+inner(f,q1)/inner(q1,q1)*q1+inner(f,q2)/inner(q2,q2)*q2
```

```python
g=proj(exp(t))
display(g)
```

```python
plot(exp(t),g,(t,0,1), legend=True)
```

```python
plot(exp(t)-g,(t,0,1))
```
