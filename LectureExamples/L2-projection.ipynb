```python
from sympy import *

t=Symbol('t')


p0=1
p1=t
p2=t**2


def inner(f,g):
    return integrate(f*g,(t,0,1))
```

```python
# Gram schmidt

q0=p0
q0norm=sqrt(inner(q0,q0))

q1=p1-inner(p1,q0)/q0norm**2*q0
q1norm=sqrt(inner(q1,q1))
q2=p2-inner(p2,q0)/

q2=p2-inner(p2,q0)/q0norm**2*q0-inner(p2,q1)/q1norm**2*q1
q2norm=sqrt(inner(q2,q2))
```

```python
display(q0)
display(q1)
display(q2)
```

```python
plot(q0/q0norm,q1/q1norm,q2/q2norm,(t,0,1))
```

```python
display(inner(q0, exp(t)))
display(inner(q1, exp(t)))
display(inner(q2, exp(t)))
```

```python
def proj(f):
    return inner(f,q0)/q0norm**2*q0+inner(f,q1)/q1norm**2*q1+inner(f,q2)/q2norm**2*q2
```

```python
g=proj(exp(t))
display(g)
g.label='g'
```

```python
plot(exp(t),g,(t,0,1))
```

```python
plot(exp(t)-g,(t,0,1))
```

```python
expand(g)# coefficients
gp=Poly(g,t)
for c in gp.all_coeffs():
    display(float(c))
```
