```python
import numpy as np
```

```python
def power_method(A, x0, N, printing=False):
    x=x0
    for _ in range(N):
        y=A.dot(x)
        mu = y[np.argmax(np.abs(y))]
        x=y/mu
        if(printing):
            print("mu=")
            print(mu)
            print("x=")
            print(x)
            print("\n")
    return x, mu
        
```

```python
#5.8.7
A=np.array([[6, 7], [8,5]])
x0=np.array([1,0])
```

```python
x, mu = power_method(A,x0, 10)
```

```python
#5.8.8
A=np.array([[2, 1], [4,5]])
x0=np.array([1,0])
```

```python
x, mu=power_method(A, x0, 10, True)
```

```python
# Power Method doesn't always converge
A= np.array([[2, -3], [3, 2]])
x0=np.array([1,0])
power_method(A,x0, 10, True)
```

```python
def power_method_Rayleigh(A, x0, N, printing=False):
    x=x0
    for _ in range(N):
        y=A.dot(x)
        mu = y[np.argmax(np.abs(y))]
        r= x.dot(y)/x.dot(x)
        x=y/mu
        if(printing):
            print("mu= "+str(mu) + "\t r= "+str(r))
            print("x=")
            print(x)
            print("\n")
    return x, mu,r
```

```python
# 5.8.11.
A=np.array([[5, 2], [2,2]])
x0=np.array([1,0])
```

```python
power_method_Rayleigh(A,x0,10, True)
```

```python
def inv_power_method(A, x0, alpha, N, printing=False):
    x=x0
    I = np.eye(A.shape[-1])
    for _ in range(N):
        y=np.linalg.solve(A-alpha*I, x)
        mu = y[np.argmax(np.abs(y))]
        lam = alpha+1/mu
        x=y/mu
        if(printing):
            print("lambda= "+str(lam))
            print("x=")
            print(x)
            print("\n")
    return x, lam
        
```

```python
# 5.8.17
A= np.array([[10, -8, -4],
             [-8, 13, 4],
             [-4, 5, 4]])
x0=np.array([1,0,0])
alpha=3.3
inv_power_method(A, x0, alpha, 5, True)
```

```python
def jacobi(A, b, x0, Niter, printing=False):
    d=np.diag(A)
    N = A-np.diag(d)
    x=x0
    for _ in range(Niter):
        x=(b-N.dot(x))/d
        if(printing):
            print(x)
    return x
```

```python
A= np.array([[5, 1, 2], 
             [-1,4,-1],
             [-2, 6, 10]])
b=np.array([1,1,1])
x0=np.zeros(3)
```

```python
jacobi(A,b, x0,10, True)
```
